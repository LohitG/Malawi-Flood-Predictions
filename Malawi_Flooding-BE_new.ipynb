## Mean of 2 consecutive weeks - 2015

```python
counter1 = 1
counter2 = 1
for col in df.columns:
    if '2015' in col and counter1 <= 7:
        new_df[f'2015_mean_{counter2}'] = (df[f'precipitation_week{counter1}_2015'] + df[f'precipitation_week{counter1+1}_2015'])/2
        counter1 += 2
        counter2 += 1
```

```python
sns.scatterplot(x='X', y='Y', data=new_df, hue='elevation')
```

## Maximum within 3 consecutive weeks - 2015

```python
maximum = 0
counter1 = 1
for col in df.columns:
    if ('2015' in col) and (counter1 <= 6):
        value = df[f'precipitation_week{counter1}_2015'] + df[f'precipitation_week{counter1+1}_2015'] + df[f'precipitation_week{counter1+2}_2015']
        maximum = np.maximum(maximum, value)
        counter1 += 1
```

```python
new_df['maximum_3weeks_2015'] = maximum
```

```python
target_variable = df['target_2015']
X_train, X_test, y_train, y_test = train_test_split(
    new_df[["X", "Y", "elevation", "Land_type", "2015_mean_1", "2015_mean_2", "2015_mean_3", "2015_mean_4", "2015_mean_5"]],   # X
     target_variable, # y
     random_state=13   # ensure reproductibility
)
```

```python
print('X_train', X_train.shape)
print('y_train', y_train.shape)
print('X_test',  X_test.shape)
print('y_test',  y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(model.intercept_)
```

```python
model.coef_
```

```python
predictions = model.predict(X_test)
X = X_test.reset_index().copy()
X["target_2015"] = y_test.tolist()
X["prediction"] = predictions
X.head()
```

```python
for y, y_pred in list(zip(y_test, predictions))[:5]:
    print("Real value: {:.3f} Estimated value: {:.5f}".format(y, y_pred))
```

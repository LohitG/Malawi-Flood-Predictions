```python
df['LC_Type1_mode'] = df['LC_Type1_mode'].apply(str)
```

```python
land_types_be = df[["LC_Type1_mode"]]
```

```python
categorical_columns = df.select_dtypes([object, "category"]).columns
```

```python
categorical_data = df[categorical_columns]
```

```python
data_be = pd.get_dummies(
                                    land_types_be, 
                                    drop_first=True,
                                    dtype="int64"
)
```

```python
df = pd.concat([
                            df,
                            data_be # reset index so it matches the numerical
                        ], axis=1
)
```
